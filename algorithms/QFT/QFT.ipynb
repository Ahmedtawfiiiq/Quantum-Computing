```python
from qiskit import QuantumCircuit
from qiskit.circuit.library import QFT
from qiskit.quantum_info import Statevector

n = 4
qc = QuantumCircuit(4)

qc.x(0)
qc.x(2)

qft = QFT(n, insert_barriers=True)
qc.compose(qft, inplace=True)

qc.decompose().draw(output="mpl")
```

```python
statevector = Statevector(qc)
statevector.draw("latex")
```

```python
from qiskit.circuit.library import QFT

n = 4
qft = QFT(n, insert_barriers=True)
iqft = QFT(n, inverse=True, insert_barriers=True)

display(qft.decompose().draw(output="mpl", reverse_bits=True))
display(iqft.decompose().draw(output="mpl", reverse_bits=True))
```
